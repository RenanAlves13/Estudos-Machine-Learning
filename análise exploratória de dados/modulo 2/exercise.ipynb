```python
# Apply a log transform to reduce the influence of very large prices.
log_transformed = np.log(dataset['Price (£)'])
sns.displot(log_transformed)
```

```python
print("Skewness: %f" % (log_transformed).skew())
```

### Interpreting the log transformation

As we can see, the log transformation made the `Price (Â£)` distribution more symmetrical and closer to a bell curve. In other words, the right tail became less stretched, which reduced the skewness and brought the data closer to the normality assumption often used in regression analysis.

There are other ways to correct skewness besides `np.log()`. Two common alternatives are the Square Root Transform (`np.sqrt`) and the Box-Cox Transform (`stats.boxcox` from `scipy`). The Box-Cox method can be especially useful when we want to search for a transformation that makes the data more normally distributed, but it only works with strictly positive values.

Reference: [Top 3 Methods for Handling Skewed Data](https://towardsdatascience.com/top-3-methods-for-handling-skewed-data-1334e0debf45)


```python
from scipy import stats

price = dataset['Price (Â£)'].dropna()

sqrt_transformed = np.sqrt(price)
boxcox_transformed, boxcox_lambda = stats.boxcox(price)

comparison = pd.DataFrame({
    'Original': [price.skew()],
    'Log': [log_transformed.skew()],
    'Square Root': [sqrt_transformed.skew()],
    'Box-Cox': [pd.Series(boxcox_transformed).skew()]
})

print(f"Box-Cox lambda: {boxcox_lambda:.4f}")
comparison
```

```python
sns.displot(price)
sns.displot(log_transformed)
sns.displot(sqrt_transformed)
sns.displot(boxcox_transformed)
```

### Practical note

- `np.log()` is a strong option when the data are right-skewed and all values are positive.
- `np.sqrt()` is usually a lighter transformation than log and may help when the skewness is moderate.
- `stats.boxcox()` is more flexible because it estimates the best power transformation automatically.
- After transforming the data, compare the new skewness and inspect the plots again before choosing which version to use in the model.

## Missing Values

The chart below confirms that this dataset does not contain missing values. The following examples still document two common cleaning strategies: removing rows with missing data and filling them with a representative statistic.

```python
# Count missing values per column and visualize the result.
total = dataset.isnull().sum().sort_values(ascending=False)
total_selected = total.head(20)
total_selected.plot(kind='bar', figsize=(8,6), fontsize=10)

plt.xlabel("Columns", fontsize=20)
plt.ylabel("Count", fontsize=20)
total
```

```python
# Example 1: if Price (£) had missing values, we could remove those rows.
dataset.dropna(subset=['Price (£)'])
```

```python
# Example 2: compute the median, a common replacement for missing numeric values.
median = dataset["Price (£)"].median()
median
```

```python
# Replace missing values without using inplace=True to avoid chained-assignment warnings.
dataset['Price (£)'] = dataset['Price (£)'].fillna(median)
```

## Feature Scaling

Scaling is useful when numeric variables have different ranges. Here we compare Min-Max normalization and standardization.

```python
# Min-Max scaling rescales each numeric feature to the [0, 1] interval.
norm_data = MinMaxScaler().fit_transform(dataset_int)
norm_data
```

```python
# Convert the NumPy array back to a DataFrame so the column names are preserved.
norm_data_frame = pd.DataFrame(
    data=norm_data,
    columns=dataset_int.columns,
    index=dataset_int.index
)
norm_data_frame
```

```python
# Standardization centers each feature at 0 and scales it to unit variance.
scaled_data = StandardScaler().fit_transform(dataset_int)
scaled_data
```

```python
# Convert the standardized values into a labeled DataFrame for inspection.
scaled_data_frame = pd.DataFrame(
    data=scaled_data,
    columns=dataset_int.columns,
    index=dataset_int.index
)
scaled_data_frame
```

### Deleting outliers

The next heuristic removes the properties with the largest Square Meters values so we can compare the scatter plot before and after trimming extreme cases.

```python
# Select the 50 largest properties as a simple outlier candidate group.
data_outliers = dataset.sort_values(by='Square Meters', ascending=False)[:50]
data_outliers.count()
```

```python
# Remove those candidate outliers using their row indexes.
outliers_dropped = dataset.drop(index=data_outliers.index)
outliers_dropped
```

```python
# Compare the trimmed dataset with the original one.
new_plot = outliers_dropped.plot.scatter(x='Square Meters',
                                         y='Price (£)')
new_plot2 = dataset.plot.scatter(x='Square Meters',
                                 y='Price (£)')
```

## Z-score Analysis

Z-score measures how many standard deviations a value is from the mean. A common rule of thumb is to investigate observations with |Z| > 3 as potential outliers.

```python
# Compute the Z-score of the price column and summarize the result.
dataset['LQFSF_Stats'] = stats.zscore(dataset['Price (£)'])
dataset[['Price (£)','LQFSF_Stats']].describe().round(3)
```
